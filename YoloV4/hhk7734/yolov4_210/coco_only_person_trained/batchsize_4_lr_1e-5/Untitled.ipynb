```python
from tensorflow.keras import callbacks, optimizers
import os
os.chdir("../..")
from yolov4_210.py_src.yolov4.tf import SaveWeightsCallback, YOLOv4
os.chdir("coco_only_person_trained/batchsize_4_lr_1e-5")
import time
import datetime
start = datetime.datetime.now()
print( start )

yolo = YOLOv4(tiny=True)
yolo.classes = "/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/Wildtrack.names"
yolo.input_size = 608
yolo.batch_size = 4

yolo.make_model(activation1="relu")
yolo.load_weights(
    "/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/yolov4-tiny.conv.29",
    weights_type="yolo"
)

train_data_set = yolo.load_dataset(
    "/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/coco_train_only_person.txt",
    image_path_prefix="/home/css-wu/CloudStation/Data/train2017",
    label_smoothing=0.05
)
val_data_set = yolo.load_dataset(
    "/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/coco_val_only_person.txt",
    image_path_prefix="/home/css-wu/CloudStation/Data/val2017",
    training=False
)

epochs = 400
lr = 1e-5

optimizer = optimizers.Adam(learning_rate=lr)
yolo.compile(optimizer=optimizer, loss_iou_type="ciou")

def lr_scheduler(epoch):
    if epoch < int(epochs * 0.5):
        return lr
    if epoch < int(epochs * 0.8):
        return lr * 0.5
    if epoch < int(epochs * 0.9):
        return lr * 0.1
    return lr * 0.01

_callbacks = [
    callbacks.LearningRateScheduler(lr_scheduler),
    callbacks.TerminateOnNaN(),
    callbacks.TensorBoard(
        log_dir="/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/yolov4_210/coco_only_person_trained/batchsize_4_lr_1e-5/logs",
    ),
    SaveWeightsCallback(
        yolo=yolo, dir_path="/home/css-wu/Documents/Machine_Learning/YoloV4/hhk7734/yolov4_210/coco_only_person_trained/batchsize_4_lr_1e-5/trained",
        weights_type="yolo", epoch_per_save=10
    ),
]

yolo.fit(
    train_data_set,
    epochs=epochs,
    callbacks=_callbacks,
    validation_data=val_data_set,
    validation_steps=50,
    validation_freq=5,
    steps_per_epoch=100,
)

end = datetime.datetime.now()
print( start )
print(end)
print("total", end - start)
```
